```python
data       
# Fiyat = Y  ,   BisikletOzellik1 ve BisikletOzellik2 = X
```

```python
Y = data["Fiyat"].values
X = data[["BisikletOzellik1","BisikletOzellik2"]].values
X_train,X_test,Y_train,Y_test = train_test_split(X , Y, test_size=0.33 , random_state=30)
# X = features , two dimensional
# Y = wx + b  => label
# .values = creates an array
# split data
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
# also we could write in one line : scaled_data = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(5,activation="relu"))  
model.add(Dense(5,activation="relu"))
model.add(Dense(5,activation="relu"))
# three hidden layers 
model.add(Dense(1))
model.compile(optimizer="rmsprop" , loss="mse")
```

```python
model.fit(X_train,Y_train,epochs=150)
```

```python
model.history.history
```

```python
loss_values = model.history.history["loss"]
```

```python
loss_values
```

```python
import matplotlib.pyplot as plt
plt.plot(loss_values)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss During Training")
plt.show()
```

```python
train_loss = model.evaluate(X_train,Y_train,verbose=0)
```

```python
test_loss = model.evaluate(X_test,Y_test,verbose=0)
```

```python
train_loss
```

```python
test_loss
```
